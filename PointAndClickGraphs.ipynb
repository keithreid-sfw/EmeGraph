```python
"""
MIT license
Dr Keith Reid 2024
Eme Graph
"""

import numpy as np
import matplotlib.pyplot as plt; plt.ion()
import networkx
import netgraph

# bit of magic
%matplotlib notebook

# ask them what they care about
labels    = []
answer    = None
while answer != "STOP":
    answer = input("Please write something important to you or write STOP\t")
    labels.append(answer)
stop       = labels.pop()
node_count = len(labels)
labels     = zip(range(node_count), labels)

# make the graph
total_nodes             = node_count
weights                 = np.random.rand(total_nodes, total_nodes)
connection_probability  = 1
is_connected            = np.random.rand(total_nodes, total_nodes) <= connection_probability
graph                   = np.zeros((total_nodes, total_nodes))
graph[is_connected]     = weights[is_connected]
g                       = networkx.from_numpy_array(graph, networkx.DiGraph)

# make it so you can change it
pos                     = networkx.layout.spring_layout(g)
plot_instance           = netgraph.InteractiveGraph(graph, 
                                                    node_labels=True,                                         node_positions=pos)
# give them a key
for i in labels:
    print(i)
```
